## Entendendo os Pedidos do E-commerce 

O comércio eletrônico está cada vez mais popular e por isso vamos explorar os registros de pedidos para examinar como essa tendência se manifesta no contexto brasileiro, ao menos dentro da amostra de dados disponível.

Ao analisarmos as diferentes colunas do conjunto de dados, é evidente que os pedidos apresentam diversas condições, cada um com status de data e hora distintos, como "approved", "delivered" e "canceled". 

Para iniciar nossa análise, vamos primeiro examinar __qual é a distribuição dos diferentes status de pedidos e suas participações relativas no conjunto de dados.__

```python
# Realizar uma junção (join) usando a coluna 'customer_id'
df_orders = pd.merge(olist_orders, olist_customer,how='left', on='customer_id') 

# Criando figura e eixos
fig, ax = plt.subplots(figsize=(12, 4))

# Cria um gráfico de contagem (histograma de contagem) da coluna 'order_status'
sns.countplot(data=df_orders, x='order_status', ax=ax, order=df_orders['order_status'].value_counts().index)

# Adiciona rótulos com quantidades numéricas e relativas em cada barra
total = len(df_orders)

for p in ax.patches:             # Itera sobre as barras (ou patches) do gráfico. Cada iteração representa uma barra no gráfico.
    
    altura = p.get_height()      # Altura da barra atual, ou seja, a contagem do valor representado por essa barra.
    porcentagem = altura / total #  Calcula a porcentagem relativa dessa barra em relação ao total de observações.
    
    # Adicionar rótulos para cada barra:    
    ax.annotate(f'{altura}', 
                (p.get_x() + p.get_width() / 2., altura),  # Define a posição e altura onde o rótulo será colocado.
                ha='center', 
                va='bottom', 
                fontsize=8, 
                color= CINZA4, 
                xytext=(0, 12),  # Ajuste horizontal e vertical da posição do rótulo.
                textcoords='offset points')  # Indica que o deslocamento é especificado em pontos.
    
    ax.annotate(f'{porcentagem:.1%}', 
                (p.get_x() + p.get_width() / 2., altura),# Define a posição e altura onde o rótulo será colocado.
                ha='center', 
                va='top', 
                fontsize=9, 
                color=AZUL2,
                xytext=(0, 10),# Ajuste horizontal e vertical da posição do rótulo.
                textcoords='offset points')  # Indica que o deslocamento é especificado em pontos.

# Remove as bordas superior e da direita do gráfico
sns.despine()

# Adiciona um título ao gráfico
ax.set_title("Quantidade total de pedidos por Status") 
ax.set_ylabel('')
ax.set_xlabel('')

plt.show()
```

_Pelo gráfico podemos observar que 97% dos pedidos possuem o status de 'delivered' e apenas 3% representa os outros status. A quantidade de pedidos cancelados tambem é bem baixa, não representando nem 1%._

Para termos uma visão completa sobre o cenário do comércio eletrônico no Brasil e sua tendência, vamos precisar criar algumas colunas e corrigir alguns dados, afim de explorar a questão temporal. Alguns questionamentos levantados sobre o conjunto de dados:

- Como os pedidos evoluem ao longo do tempo?
- Existe periodos sazonais?
- Quais as participações das regiões sobre os pedidos? 
- Quais periodos do dia ocorrem maior movimentação?


```python
# Alterando oos tipos de dados para as colunas com data
timestamp_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
                  'order_delivered_customer_date','order_estimated_delivery_date']
for col in timestamp_cols: 
    df_orders[col] = pd.to_datetime(df_orders[col]) # muda para datetime as colunas da lista 'timestamp_cols'.
```

```python
# Criar colunas para divisão temporal necessária.

# Obtendo os atributos de Ano e Mês para data de compra
df_orders['order_purchase_year'] = df_orders['order_purchase_timestamp'].apply(lambda x: x.year)
df_orders['order_purchase_month'] = df_orders['order_purchase_timestamp'].apply(lambda x: x.month)
df_orders['order_purchase_month_name'] = df_orders['order_purchase_timestamp'].apply(lambda x: x.strftime('%b'))
df_orders['order_purchase_year_month'] = df_orders['order_purchase_timestamp'].apply(lambda x:x.strftime('%Y%m'))
df_orders['order_purchase_date'] = df_orders['order_purchase_timestamp'].apply(lambda x:x.strftime('%Y%m%d'))

# Obtendo os atributos de dia e dia da semana para data de compra
df_orders['order_purchase_day'] = df_orders['order_purchase_timestamp'].apply(lambda x: x.day)
df_orders['order_purchase_dayofweek'] = df_orders['order_purchase_timestamp'].apply(lambda x:x.dayofweek)
df_orders['order_purchase_dayofweek_name'] = df_orders['order_purchase_timestamp'].apply(lambda x:x.strftime('%a'))

# Obtendo os atributos de hora e periodo do dia para data de compra
df_orders['order_purchase_hour'] = df_orders['order_purchase_timestamp'].apply(lambda x:x.hour)
hours_bins = [-0.1, 6, 12, 18, 23] # lista para definir os limites dos intervalos para categorizar
hours_labels = ['Overnight', 'Morning', 'Afternoon', 'Night'] # lista com os nomes dos intervalos 
df_orders['order_purchase_time_day'] = pd.cut(df_orders['order_purchase_hour'], hours_bins, labels=hours_labels)
```

```python
# Criar coluna com a diferença de dias da data do pedido e a aprovação no df 'filt_df_orders'
filt_df_orders['diff_days_approved_customer'] = (filt_df_orders['order_approved_at'] - filt_df_orders['order_purchase_timestamp']).dt.days

# Criar uma coluna com a quantidade de dias para a entrega para a distribuidora
# (order_delivered_customer_date - order_purchase_timestamp)
filt_df_orders['diff_days_timestamp_customer'] = (filt_df_orders['order_delivered_customer_date'] - filt_df_orders['order_purchase_timestamp']).dt.days

# Criar uma coluna com a quantidade de dias para a entrega para a distribuidora
# (order_delivered_carrier_date - order_purchase_timestamp)
filt_df_orders['delivered_diff_time_days'] = (filt_df_orders['order_delivered_carrier_date'] - filt_df_orders['order_purchase_timestamp']).dt.days

# Criar uma coluna com a quantidade de dias para a entrega da transportadora até o cliente
# (order_delivered_customer_date - order_delivered_carrier_date)
filt_df_orders['diff_days_carrier_customer'] = (filt_df_orders['order_delivered_customer_date'] - filt_df_orders['order_delivered_carrier_date']).dt.days

# Criar coluna com a diferença de dias da data entregue até a previsão de entrega
# (order_estimated_delivery_date - order_delivered_customer_date)
filt_df_orders['diff_days_estimated_customer'] = (filt_df_orders['order_estimated_delivery_date'] - filt_df_orders['order_delivered_customer_date']).dt.days
```

No gráfico Boxplot são considerados valores discrepantes(Outliers) quando o valor da variável (VAR):

- **VAR > Q3 + 1.5IQR**
- **VAR < Q1 + 1.5IQR**

E o que são Q1 (primeiro quartil), Q3 (terceiro quartil) e IQR (Distância Interquartil)?

_**Q1**: Representa o ponto que divide os 25% menores valores da coluna, também conhecido como primeiro quartil._

_**Q3**: Indica o valor que separa os 25% maiores valores da coluna, também chamado de terceiro quartil._

_**IQR** = Q3 - Q1_

Por se tratar de um E-commerce, os outliers podem indicar que alguns clientes estão propensos a gastar mais com determinados produtos, devido ao ticket alto de compra. Este cenário abre outras possibilidades de análises, entretanto seguiremos afim de compreender os dados em geral.

```python
# Criando figura
fig = plt.figure(constrained_layout=True, figsize=(12, 10))

# Definindo eixo
gs = GridSpec(2, 2, figure=fig, hspace=0.1)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, :2])

# Cria um gráfico de contagem (histograma de contagem) da coluna 'payment_type'
sns.countplot(data=df_orders_payments, x='payment_type', ax=ax1,palette='YlGnBu_r',
              order=df_orders_payments['payment_type'].value_counts().index)

# Adiciona rótulos com quantidades numéricas e relativas em cada barra
total = len(df_orders_payments)

for p in ax1.patches:            # Itera sobre as barras (ou patches) do gráfico. Cada iteração representa uma barra no gráfico.
    
    altura = p.get_height()      # Altura da barra atual, ou seja, a contagem do valor representado por essa barra.
    porcentagem = altura / total #  Calcula a porcentagem relativa dessa barra em relação ao total de observações.
    
    # Adicionar rótulos para cada barra:  
    ax1.annotate(f'{porcentagem:.1%}', 
                (p.get_x() + p.get_width() / 2., altura),# Define a posição e altura onde o rótulo será colocado.
                ha='center', 
                va='top', 
                fontsize=9, 
                color=CINZA4, 
                xytext=(0, 10),# Ajuste horizontal e vertical da posição do rótulo.
                textcoords='offset points')  # Indica que o deslocamento é especificado em pontos.

# Rótulos
ax1.set_title('Transações por Tipo de Pagamento', color = CINZA3,fontweight='bold')
ax1.set_xlabel('')
ax1.set_ylabel('')  
sns.despine(ax=ax1)

# Cria um gráfico de contagem (histograma de contagem) da coluna 'payment_installments'
sns.countplot(data=df_orders_payments, y='payment_installments', ax=ax2,palette='GnBu_r',
              order=df_orders_payments['payment_installments'].value_counts().index)

# Adiciona rótulos com quantidades numéricas e relativas em cada barra
total = len(df_orders_payments)

for p in ax2.patches:
    altura = p.get_width()
    porcentagem = (altura / total) * 100

    # Adicionar rótulos para cada barra:
    ax2.annotate(f'{int(altura)}',
                 (altura, p.get_y() + p.get_height() / 2),
                 ha='left', 
                 va='center', 
                 fontsize=10, 
                 color=CINZA4, 
                 xytext=(5, 0), 
                 textcoords='offset points')

    ax2.annotate(f'{porcentagem:.1f}%', 
                 (altura, p.get_y() + p.get_height() / 2),
                 ha='left', 
                 va='center', 
                 fontsize=10, 
                 color= AZUL3, 
                 xytext=(40, 0), 
                 textcoords='offset points')
    
# Definir rótulos e título
ax2.set_xlabel('')
ax2.set_ylabel('Parcelas de Pagamento', color = CINZA4)
ax2.set_title('Distribuição das Parcelas de Pagamento', color=CINZA3,fontweight='bold')

# Desliga o eixo x
ax2.get_xaxis().set_visible(False)
sns.despine(bottom=True, ax=ax2)

#agrupar
payment_agg = df_orders_payments.groupby(['order_purchase_year_month','payment_type']).size().reset_index(name='count_payment')

# Defina a lista de cores
cores = ['cornflowerblue', 'navy', 'royalblue', 'lightsteelblue']

sns.lineplot(data=payment_agg, x='order_purchase_year_month', y='count_payment', hue='payment_type', palette=cores, ax=ax3)

# Adicionar legenda na frente de cada linha
for payment, data in payment_agg.groupby('payment_type'):
    x = data['order_purchase_year_month']
    y = data['count_payment']
    cor = cores[payment_agg['payment_type'].unique().tolist().index(payment)]  # Obtém a cor da lista
    ax3.text(x.iloc[-1], y.iloc[-1], payment, fontsize=11, va='center', ha='left', color=cor)

ax3.set_title('Volume de Transações por Tipo de Pagamento ao Longo do Tempo', color = CINZA3,fontweight='bold')
ax3.set_xlabel('')
ax3.set_ylabel('')      

# Destacar parte do gráfico
ax3.fill_between(x=[16, 19], y1=0, y2=7000, color='lightgrey', alpha=0.3, zorder=0.5)

# Ajustar os titulos do eixo X
for tick in ax3.get_xticklabels():
    tick.set_rotation(45)

# Título Geral
fig.text(0, 1.08, 'Composição de Pagamentos no Comércio Online Brasileiro', 
         fontsize=16, 
         color=CINZA3, 
         fontweight='bold')

# Subtítulo Geral
fig.text(0, 1.035, 'Uma visão geral das preferências de pagamentos do público brasileiro no E-Commerce. ' 
                   'Destaque para o cartão de crédito e pagamentos em até \n10x, que também são uma preferência do consumidor. '
                   'Em maio de 2018, pagamentos com cartão de débito ensaiam uma tendência de alta.',
         fontsize=12, 
         color=CINZA5)

# Definir cor cinza dos valores do eixo y
ax1.tick_params(axis='y', colors=CINZA3, labelsize=10)
ax2.tick_params(axis='y', colors=CINZA3, labelsize=10)
ax3.tick_params(axis='y', colors=CINZA3, labelsize=10)

# Definir cor cinza dos valores do eixo x
ax1.tick_params(axis='x', colors=CINZA3)
ax2.tick_params(axis='x', colors=CINZA3)
ax3.tick_params(axis='x', colors=CINZA3)    
    
# Remover a caixa da legenda
ax3.get_legend().remove()    
sns.despine(ax=ax3)
plt.show()
```

Podemos observar grande predominância do cartão de crédito e do boleto como forma de pagamento, que segue as tendências anteriores. Dos consumidores com IDs duplicados no dataframe, esses casos envolvem clientes que efetuaram múltiplas compras, sendo que a duplicidade ocorreu devido à agregação de dois dataframes. Esse fenômeno é resultado tanto do order_item_id quanto de pedidos que incluem mais de um produto.

Diante disso, faremos ajustes nos dados para uma análise mais aprofundada desses clientes, visando compreender de maneira mais eficaz suas interações e comportamentos.

```python
# Qual o grupo de clientes que realizou a segunda compra?

df_second_purchase = df_orders_sellers.copy() # dataframe considera de 012017 a 082018 e pedidos entregues

# Ordenar o dataframe por customer_unique_id e order_purchase_timestamp
df_second_purchase.sort_values(by=['customer_unique_id', 'order_purchase_timestamp'], inplace=True)
```

```python
df_second_purchase
```

```python
# Identificar o numero da compra do cliente que a movimentação se refere.

# Inicializando a nova coluna 'purchase_number' com 1
df_second_purchase['purchase_number'] = 1

for i in range(1, len(df_second_purchase)):
    # Verificando se o 'customer_unique_id' é o mesmo da linha anterior
    if df_second_purchase['customer_unique_id'].iloc[i] == df_second_purchase['customer_unique_id'].iloc[i-1]:
        # Verificando se a 'order_purchase_timestamp' é diferente da linha anterior
        if df_second_purchase['order_purchase_timestamp'].iloc[i] != df_second_purchase['order_purchase_timestamp'].iloc[i-1]:
            # Incrementando o 'purchase_number' se a 'order_purchase_timestamp' for diferente
            df_second_purchase['purchase_number'].iloc[i] = df_second_purchase['purchase_number'].iloc[i-1] + 1
        else:
            # Mantendo o 'purchase_number' igual ao da linha anterior se a 'order_purchase_timestamp' for a mesma
            df_second_purchase['purchase_number'].iloc[i] = df_second_purchase['purchase_number'].iloc[i-1]
    else:
        # Reiniciando o 'purchase_number' quando 'customer_unique_id' é diferente
        df_second_purchase['purchase_number'].iloc[i] = 1             
```

```python
# contagem de quantos clientes têm quantas compras únicas.
df_second_purchase['customer_second_purchase'] = df_second_purchase.groupby('customer_unique_id')['purchase_number'].transform('nunique')
df_second_purchase['customer_second_purchase'].value_counts()
```

```python
# Identificar clientes que compraram mais de uma vez
df_second_purchase['customer_second_purchase'] = (df_second_purchase['customer_second_purchase'] > 1).astype(int)
```

```python
# Calcular a diferença entre a primeira compra e a segunda compra
df_filtro['primeira_compra'] = df_filtro.groupby('customer_unique_id')['order_purchase_timestamp'].transform('first')
df_filtro['segunda_compra'] = df_filtro.groupby('customer_unique_id')['order_purchase_timestamp'].transform(lambda x: x.iloc[1] if len(x) > 1 else None)

# Converter as colunas para o formato datetime
df_filtro['primeira_compra'] = pd.to_datetime(df_filtro['primeira_compra'])
df_filtro['segunda_compra'] = pd.to_datetime(df_filtro['segunda_compra'])

# Calcular a diferença entre as compras em dias
df_filtro['dif_dias'] = (df_filtro['segunda_compra'] - df_filtro['primeira_compra']).dt.days

# Exiba o DataFrame resultante
df_filtro
```
